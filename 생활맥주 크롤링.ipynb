```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get("https://dailybeer.co.kr/board/index.php?board=menu_01&sca=2") # 생활맥주 수제맥주 페이지
```

```python
html = req.text
```

```python
soup_obj = BeautifulSoup(html, 'html.parser')
```

```python
info_ttl = soup_obj.select("p.info_ttl_kor")

beer_detail = soup_obj.select("p.text.beer_detail")

#bg_image = soup_obj.select("") #이미지와 비디오 크롤링 코드 다시 짜기

#video_bg = soup_obj.select("")
```

```python
beer_menu = [] # 빈 리스트를 전역변수로 생성
for eachLine in zip(info_ttl, beer_detail):
	beer_menu.append(
        {
            eachLine[0].text:eachLine[1].text #이름과 디테일한 설명 리스트로 저장
            
        }
) 
for i in beer_menu:
    print(i)
```
